#### Goal : In addition to Start and Stop buttons, add two more buttons: Pause and Resume

```python
import tkinter as tk
import time
import threading
from tkinter import ttk

root = tk.Tk()
root.title("Assign5_Buttons")

root.geometry("200x420")

stop_thread = False
pause_thread=False

progressbar = ttk.Progressbar(root, orient = "vertical", length=200, mode = "determinate")
progressbar.pack()

label = tk.Label(root, text = "0%")
label.pack()


def start_progress():
    for i in range(101):
        global stop_thread
        if stop_thread:
            break
        global pause_thread
        while pause_thread and not stop_thread: 
            time.sleep(0.1)
        progressbar['value']=i
        label.config(text=f"{i}%")
        time.sleep(0.1)
        root.update()

def start():
    global stop_thread, pause_thread
    stop_thread = False
    pause_thread = False
    progressbar["value"]=0
    label.config(text=f"0%")
    t = threading.Thread(target = start_progress)
    t.start()
    start_button.config(state=tk.DISABLED)
    stop_button.config(state=tk.NORMAL)
    pause_button.config(state=tk.NORMAL)
    resume_button.config(state=tk.DISABLED)

    

def stop():
    global stop_thread
    stop_thread = True
    progressbar["value"]=0
    label.config(text="0%")
    t = threading.Thread(target = start_progress)
    t.start()
    start_button.config(state=tk.NORMAL)
    stop_button.config(state=tk.DISABLED)
    pause_button.config(state=tk.DISABLED)
    resume_button.config(state=tk.DISABLED)

def pause():
    global pause_thread
    pause_thread=True
    start_button.config(state=tk.DISABLED)
    stop_button.config(state=tk.NORMAL)
    pause_button.config(state=tk.DISABLED)
    resume_button.config(state=tk.NORMAL)

def resume():
    global pause_thread
    pause_thread = False
    start_button.config(state=tk.DISABLED)
    stop_button.config(state=tk.NORMAL)
    pause_button.config(state=tk.NORMAL)
    resume_button.config(state=tk.DISABLED)
    

start_button = tk.Button(root, text="Start", command= start)
start_button.pack(pady=10)
stop_button = tk.Button(root, text="Stop", command = stop)
stop_button.pack(pady=10)

pause_button =tk.Button(root, text="Pause", command = pause)
pause_button.pack(pady=10)
resume_button = tk.Button(root, text="Resume", command = resume)
resume_button.pack(pady=10)

stop_button.config(state=tk.DISABLED)
pause_button.config(state=tk.DISABLED)
resume_button.config(state=tk.DISABLED)

root.mainloop()
```
